```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie=pd.read_csv('movies_metadata.csv')
```

```python
movie.head()
```

```python
cred=pd.read_csv('credits.csv')
```

```python
cred.head()
```

```python
df=pd.DataFrame()
```

```python
def get_name(x):
    l=[]
    p=eval(x)
    for i in range(len(p)):
        l.append(p[i]['name'])
    return l
```

```python
df['adult']=movie['adult']
df['title']=movie['title']
df['genres']=movie['genres'].apply(lambda x:get_name(x))
df['cast']=cred['cast'].apply(lambda x:get_name(x))
df['director']=cred['crew'].apply(lambda x:get_name(x))
df['index']=df.index
```

```python
df
```

```python
df.columns
```

```python
feat=['adult','title','genres','cast','director']
for f in feat:
    df[f]=df[f].fillna('')
```

```python
def combfeat(row):
    return row['adult']+" "+row['title']+" "+row['genres']+" "+row['cast']+" "+row['director']
```

```python
df['comb']=df.apply(combfeat,axis=1)
```

```python
df['genres']=df['genres'].apply(lambda x: ' '.join(map(str, x)))
```

```python
df['cast']=df['cast'].apply(lambda x: ' '.join(map(str, x)))
```

```python
df['director']=df['director'].apply(lambda x: ' '.join(map(str, x)))
```

```python
df
```

```python
cv=CountVectorizer()
```

```python
count_matrix=cv.fit_transform(df['comb'])
```

```python
count_matrix
```

```python
cosin_sim=cosine_similarity(count_matrix)
```

```python
cosin_sim
```

```python
def get_title_from_index(t):
    return df[df.index==t]['title'].values[0]
```

```python
def get_index_from_title(i):
    return df[df.title==i]['index'].values[0]
```

```python
movie_like="Avatar"
```

```python
movie_index= get_index_from_title(movie_like)
```

```python
rec_movies=list(enumerate(cosin_sim[movie_index]))
```

```python
sort_mov=sorted(rec_movies,key=lambda x: x[1],reverse=True)
```

```python
print('movies for you')
for i in sort_mov:
    print(get_title_from_index(i[0]))
```
